<h1 style="font-size:25px; color:orange; text-align:center;">
    Q1- What are the airlines in the datasets , accompanied by the frequencies ?
</h1>


```python
data.head()
```

```python
# checking the how many airlines are there in the data set
data['airline'].nunique()
```

```python
# showing the unique airlines in the data set
data['airline'].unique()
```

```python
# showing the frequencies of the airlines in the data set
data['airline'].value_counts()
```

```python
# showing all the airlines with their frequencies in a bar plot
data['airline'].value_counts(ascending=True).plot.barh(color=['skyblue','lightgreen'])
plt.title('Airlines with their frequencies')
plt.xlabel('No. of Flights')
plt.ylabel('Airlines')
plt.show()
```

<h1 style="font-size:25px; color:orange; text-align:center;">
    Q2 - Show Bar Graphs Representing the Departure Time and Arrival Time ?
</h1>


```python
data.head()
```

```python
# Showing the Departure_time 
data['departure_time'].value_counts()
```

```python
# Showing the Arrival_etime 
data['arrival_time'].value_counts()
```

```python
# showing the Departure time & Arrival time with their frequencies in a bar plot

plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(data['departure_time'].value_counts().index,data['departure_time'].value_counts(),color=['gray','orange'])
plt.xlabel('D.time')
plt.ylabel('D.Frequencies')
plt.title('Departure Time with Frequencies')
plt.subplot(1,2,2)
plt.bar(data['arrival_time'].value_counts().index,data['arrival_time'].value_counts(),color=['lightblue','pink'])
plt.xlabel('A.time')
plt.ylabel('A.Frequencies')
plt.title('Arrival Time with Frequencies')


plt.show()
```

<h1 style="font-size:25px; color:orange; text-align:center;">
    Q4 - Does Price Varies with the Airlines ?
</h1>


```python
data.head()
```

```python
# Grouping the airlines and checking their mean Price

data.groupby('airline')['price'].mean()
```

```python
# Drwaing the categorial plot shwoing the price varies with airlines
sns.barplot(x='airline', y='price', data=data, palette='Set3',hue='class')
plt.title('Price Varies with Airlines')
plt.xlabel('Airlines')
plt.ylabel('Price')
plt.show()
```
